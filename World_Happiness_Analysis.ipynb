# üåç Happiness Data Analysis

This project analyzes the happiness levels of 8 countries using a small CSV dataset.  
We explore the relationships between happiness and factors like GDP and life expectancy.  
Visualizations help understand trends and rankings.

  Focus:
- Correlation between variables
- Country ranking by happiness
- Simple conclusions


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style='whitegrid')
df = pd.read_csv('happiness_sample.csv')
df.head()
```

 Dataset Overview

The dataset includes happiness scores and related factors for 8 countries.


```python
df.info()
df.describe()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
plt.bar(df['Country'],df['Happiness_Score'],color='skyblue')
plt.xlabel('Country')
plt.ylabel('Happiness_Score')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.scatter(df['GDP_per_Capita'],df['Happiness_Score'],color='green')
plt.xlabel('GDP per Capita')
plt.ylabel('Happiness Score')
plt.title('Happiness vs GDP per Capita')
plt.grid(True)
plt.show()
```

```python
df.corr(numeric_only=True)
```

```python
import seaborn as sns
plt.figure(figsize=(8, 6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm', center=0)
plt.title('Correlation Heatmap')
plt.show()
```

```python
df_sorted = df.sort_values(by='Happiness_Score',ascending=False)
plt.figure(figsize=(10, 6))
sns.barplot(x='Happiness_Score', y= 'Country',data=df_sorted, palette='viridis')
plt.title('Happiness Score by Country')
plt.xlabel('Happiness Score')
plt.ylabel('Country')
plt.show()
```

- Countries with higher GDP tend to have higher happiness scores.
